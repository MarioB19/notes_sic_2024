## Entendiendo a las derivadas


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2

def f_p(f,x,d_x=1e-5):
    return (f(x+d_x)-f(x))/d_x

def tangente(x):
    return (slope*(x-a)+fa)

a = 1.5

slope = f_p(f,a)

fa = f(a)

x_values = np.linspace(-2,3,100)
y_values= f (x_values)
tangent_values = tangente(x_values)

#Graficamos con plt

plt.figure(figsize=(10,6))
plt.plot(x_values,y_values,label='f(x) = x**2',color="blue")
plt.plot(x_values,tangent_values, "--",label = f"Tangente en {a}", color = "red")
plt.scatter([a],[fa], color = "black")
plt.text(a,fa, f"({a},{fa})", ha = "right", color = "black")
plt.xlabel("X")
plt.ylabel("f(x)")
plt.title("Derivada y relacin con la recta tangente")
plt.legend()
plt.grid(True)
plt.show()



print(slope)
```
